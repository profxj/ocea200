## We derived in class, from Sverdrup theory:

## $\beta V_S = \frac{1}{\rho_0} \left [ \frac{\partial \tau_y}{\partial x} - \frac{\partial \tau_x}{\partial y} \right ] = f w_E$

## Here, the wind stress is 

## $X = \tau_x = X_0 \sin(\pi y/L)$

### so we have

## ${\rm Curl} \, X = -\frac{\partial \tau_x}{\partial y} = (X_0 \pi / L) \cos(\pi y / L)$

## This is maximal at $y=0$, as is apparent in the numerical outputs and vanishes at the boundaries ($|y| = L/2$)

## Considering the magnitude at $y=0$, 

## $V_S = \frac{X_0 \pi}{\beta \rho_0 L}$

```python
# Pluggin in
X_0 = 0.05 * unyt.N / unyt.m**2
L = 5000 * unyt.km
beta = 2e-11 / unyt.m / unyt.s # mid-latitude estimate
rho0 = 1025 * unyt.kg / unyt.m**3
```

```python
V_S = X_0 * np.pi / beta / rho0 / L
V_S.to('m**2/s')
```
