## After 5 years, the low latitudes now show the steady state flow as does regions close to the eastern bondary at higher latitudes.  Again, this follows from the latitude dependence of the group velocity $c_g \propto f^{-2}$ of the Rossby waves.

## Anderson & Gill (1975) derive a solution for the planetary waves that arise which have group velocity:

## $c_g = \frac{k^2 - \Lambda}{(k^2 + \Lambda^2)^2}$

## For $k=0$, $c_g = -1/\Lambda$ and we have western propagation for the longest wavelength modes

## For $k > \sqrt{\Lambda}$, $c_g > 0$ with a maximum of $1/(8\Lambda)$ and the waves propagate to the East, with shorter wavelengths

## So, we have two moving boundaries within the interior with the one from the Eastern boundary moving at 8x the speed.   Meanwhile, the size of the western boundary current grows much more slowly owing to the slower $c_g$ of the East progagating waves.

## Regarding the western boundary current, Anderson and Gill (1975) emphasize (in Section 7) that the motions are dominated by the short-wave approximation yielding:

### $u_{xxt} + u_x = 0$

## which yields solutions in terms of the Bessel function $J_1$ and explains the complex velocity field near the Western boundary

## We derived in class, from Sverdrup theory:

## $\beta V_S = \frac{1}{\rho_0} \left [ \frac{\partial \tau_y}{\partial x} - \frac{\partial \tau_x}{\partial y} \right ] = f w_E$

## Here, the wind stress is 

## $X = \tau_x = X_0 \sin(\pi y/L)$

### so we have

## ${\rm Curl} \, X = -\frac{\partial \tau_x}{\partial y} = -(X_0 \pi / L) \cos(\pi y / L)$

## This is maximal at $y=0$, as is apparent in the numerical outputs and vanishes at the boundaries ($|y| = L/2$) but is negative everywhere (i.e. towards the equator).

## Considering the magnitude at $y=0$, 

## $V_S = \frac{X_0 \pi}{\beta \rho_0 L}$

```python
# Pluggin in
X_0 = 0.05 * unyt.N / unyt.m**2
L = 5000 * unyt.km
beta = 2e-11 / unyt.m / unyt.s # mid-latitude estimate
rho0 = 1025 * unyt.kg / unyt.m**3
```

```python
V_S = X_0 * np.pi / beta / rho0 / L
V_S.to('m**2/s')
```

## I think this matches (roughly) the contours on $\psi_M$

## As discussed in class, the flow leads to a decrease in $PV$:

## $PV_1 = f_1/H$ for $y>0$ and decreases to $PV_2 = f_2/H$ for $y<0$

## For $\beta = 2 \times 10^{11} \, \rm m^{-1} s^{-1}$ and $\Delta y = 5000$km, $f_2 = f_1 - 10^{-4}\,\rm s^{-1}$

## This has to be compensated by an increase in $PV$ in the flow along the western boundary current.  And we have interpreted this as originating from the divergence of the horizontal flux apparent in the $V$ field at year 50.

# (c) Consider $\psi_S$, the stream function

## Why can the boundary condition $\psi_S = 0$ be applied only at the Eastern boundary?

## Because the wind is imparting *negative* $PV$ on the water that flows towards the equator, to conserve total $PV$ we must have a source of *positive* $PV$.  This can only occur along the eastern boundary where we can enforce positive $\partial \psi_S / \partial x$ from a boundary condition of $\psi_S = 0$.
