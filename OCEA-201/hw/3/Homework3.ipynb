```python
def contour_plot(field, title=None, cmap='jet', nlevels=8, clabel=True, ax=None):
    cm = plt.get_cmap(cmap)
    if ax is None:
        fig = plt.figure(figsize=(10,10))
        ax = plt.gca()

    CS = ax.contour(field, levels=nlevels, cmap=cm)

    if clabel:
        ax.clabel(CS, inline=True, fontsize=12)
    if title is not None:
        ax.set_title(title, fontsize=18.)

    ax.set_aspect('equal', 'datalim')
```

```python
def contourf_plot(we):
    fig = plt.figure(figsize=(10,10))
    cm = plt.get_cmap('jet')

    img = plt.contourf(we, cmap=cm, levels=20)
    #
    cb = plt.colorbar(img, pad=0.)
    clbl=r'$w_E$ (m/s)'
    cb.set_label(clbl, fontsize=20.)
    cb.ax.tick_params(labelsize=17)
    #
    plt.axes().set_aspect('equal', 'datalim')
```
