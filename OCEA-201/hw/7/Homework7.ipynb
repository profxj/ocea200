```python
# imports
from importlib import reload

import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
sns.set_style('whitegrid')
sns.set_context('notebook')
```

```python
def set_fontsize(ax,fsz):
    '''
    Parameters
    ----------
    ax : Matplotlib ax class
    fsz : float
      Font size
    '''
    for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
                ax.get_xticklabels() + ax.get_yticklabels()):
        item.set_fontsize(fsz)
```

# Problem 1a

```python
period_1 = [6.1, 3.1, 2.]
efold_1 = [25., 13., 8.5]
#
period_2 = [11., 5.7, 3.8]
efold_2 = [24., 12., 7.6]
#
period_3 = [4., 2., 1.3]
efold_3 = [45., 23., 15.]
```

```python
U0 = [0.5, 1., 1.5] # m/s
```

## Plot

```python
fsz=15.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,7))
# Period
ax1.plot(U0, period_1, 'o', label='Mode 1')
ax1.plot(U0, period_2, 'o', label='Mode 2')
ax1.plot(U0, period_3, 'o', label='Mode 3')
#
ax1.set_xlabel(r'$U_0$ (m/s)')
ax1.set_ylabel('Period (days)')
#
ax1.legend()
#
# e-fold
ax2.plot(U0, efold_1, 'o', label='Mode 1')
ax2.plot(U0, efold_2, 'o', label='Mode 2')
ax2.plot(U0, efold_3, 'o', label='Mode 3')
#
ax2.set_xlabel(r'$U_0$ (m/s)')
ax2.set_ylabel('e-fold (days)')
#
ax2.legend()
#
for ax in [ax1, ax2]:
    set_fontsize(ax, fsz)
#
plt.show()
```
